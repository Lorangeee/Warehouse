```python
import numpy as np

def initialize_Weights(dims1,dims2):
  ''' this function creates random initialized vector for w with shape (dims1,dims2),and b to zero.

  Argument:
  dims1, the number of l-th layer's nodes, a scalar
  dims2, the number of l-1-th layer's nodes, a scalar  
  
  Returns:
  w - randomly initialized vector of shape (dims1,dims2)
  b - initialized to be 0
  '''

  w = np.random.randn(dims1,dims2)
  b = 0

  return w,b


```
